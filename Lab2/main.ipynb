```r
#EuStockMarkets
class(EuStockMarkets)
df <-as.data.frame(EuStockMarkets) 
df$Date <- as.numeric(time(EuStockMarkets))

#create a line chart for df 
ggplot(df, aes(x=Date, y=DAX)) + geom_line()

#create a line chart for df from 1995 to 2000
ggplot(df[df$Date>=1997,], aes(x=Date, y=DAX)) + geom_line()
```

```r
rownames(economics)
colnames(economics)
```

```r
dfEconomics <- as.data.frame(economics) 
#convert the date to only show the year
dfEconomics$date <- as.numeric(substr(dfEconomics$date, 1, 4))
#convert the saving rate from foat to percentage(12.5 = 12.5%)
dfEconomics$psavert <- as.numeric(dfEconomics$psavert/10)

#create a bar chart for dfEconomics
ggplot(dfEconomics, aes(x=date, y=psavert)) + geom_bar(stat="identity")


#create a bar chart for saving rate over time and color the bars based on saving rate(< 5% = red, > 5% = blue, >10% = green)
ggplot(dfEconomics, aes(x=date, y=psavert, fill = psavert * 10)) +
  geom_bar(stat="identity") +
  labs(x = "Date",
       y = "Saving Rate in percentage",
       fill = "Saving Rate",
       title = "Saving Rate Over Time") +
  scale_color_manual(
       values=c("red", "blue", "green"))
```

```r
#?HairEyeColor
data(HairEyeColor)
#head(HairEyeColor)

df <-  as.data.frame(HairEyeColor)
#create a dataframe for HairEyeColor males

males <-  as.data.frame(HairEyeColor[,,'Male'])
females <- as.data.frame(HairEyeColor[,,"Female"])
rownames(males)
colnames(males)


ggplot(males, aes(x=Eye, y=Freq, fill = Eye)) +
  geom_bar(stat="identity") +
    labs(
       title = "Male eye colour frequency")

#create a pie chart for hair colour 
ggplot(females, aes(x="", y=Hair, fill=Hair)) +
  geom_bar(stat="identity") +
  labs(
       title = "Female hair colour frequency") +
  coord_polar("y", start=0) + theme_void()
```

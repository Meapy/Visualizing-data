```r
library(tidyverse)
library(reshape2)
library(plotly)
```

```r
titanicdata = "data/bigtitanic.csv"
titanicdata <- read.csv(titanicdata, header=TRUE,stringsAsFactors = FALSE)
titanicdata$survived<-ifelse(titanicdata$survived==0,"No","Yes")
head(titanicdata)
```

```r
titanicdata$pclass <- as.factor(titanicdata$class)
titanicdata$survived <- as.factor(titanicdata$survived)
titanicdata$sex <- as.factor(titanicdata$sex)
titanicdata$embarked <- as.factor(titanicdata$embarked)
titanicdata = titanicdata[c(1,2,4,5)]

head(titanicdata)
```

```r
ggplot(titanicdata, aes(x = class, fill = survived)) + 
  theme_bw() +
  geom_bar() +
  labs(y = "Passenger Count",
       title = "Titanic Survival Rates by class")
```

```r
#histogram to show surivate rate by age, class and gender
ggplot(titanicdata, aes(x = age, fill = survived)) +
  theme_bw() +
  facet_wrap(sex ~ class) +
  geom_histogram(binwidth = 5) +
  labs(y = "Total number of people",
       x = "Age of the people",
       title = "Titanic Survival Rates by Age, class and Sex")
```
